## Problem Statement

Phishing remains the most prevalent entry vector for data breaches worldwide. The reality of modern cybersecurity is that Security Operations Center (SOC) analysts spend hours every single day manually reading and classifying potentially harmful emails. It is repetitive, prone to human error, and struggles against the volume and speed of modern engineered phishing attacks.

'The Hunter' solves this by automating the initial triage. When an email arrives, it goes through our deep learning ensemble model for classification. We built an autonomous AI agent security team, complete with different roles, that take that classification, reason through the email's intent, deeply analyze the threat, and automatically log the findings for the human team.

---

## Architecture & Deep Learning Integration

To make this system effective, we needed the accuracy of Neural Networks paired with the reasoning power of Large Language Models.

- **Input:** Raw email enters the system.
- **Deep Learning Model:** The classification backbone is a stacked ensemble: a Bidirectional LSTM (BiLSTM) text model weighted at 65% combined with a Logistic Regression feature model at 35%. Both are trained on the 28,747-email Alam Kaggle dataset. These understand the context and sequential patterns of the words, rather than just doing simple keyword matching.
- **Agent Brain:** Once the ensemble model spots malicious intent and gives it a risk score based on a precision-recall calibrated threshold, that score is handed off to our Agent Framework (CrewAI powered by Meta LLaMA 3 8B).
- **Tool / Output:** An Orchestrator routes tasks between our specific agents in a sequential pipeline to generate the final response and SOC threat memory via SQLite.

## The Thinking Demo & 7 Live App UI Walkthrough

In this demo, you will see the full pipeline processing demonstration emails through our intuitive Gradio UI:

## Challenges & Reflection

**Challenges we faced:**
The biggest challenge was getting the deep learning output to cleanly format and pass into the agent framework's prompt without confusing the LLM. Maintaining consistency between the predictive model and the generative roles required careful orchestration.

**What we learned:**
The most surprising thing was how much **prompt engineering** goes into building an agent persona. If you do not give the agents strict boundaries, they tend to over-analyze or skip steps! Strict roles and a sequential process - chosen over a hierarchical one - ensure smooth linear dependencies and handoffs.
